```python
%%writefile Latas.cs

using System;

namespace myApp 
{
    class Program 
    {
        static void Main()
        {
            //Declarando variáveis
            double largura_parede, altura_parede, area_parede, quantidade_tinta, quantidade_latas;

            //Pegando os dados da parede com o usuário
            Console.WriteLine("Por favor, digite a altura e largura da parede a ser pintada");
            altura_parede  = double.Parse(Console.ReadLine());
            largura_parede = double.Parse(Console.ReadLine());

            //Calculando a área
            area_parede = altura_parede * largura_parede;

            //Calculando quantidade de tinta em litros
            quantidade_tinta = area_parede * 0.3 ;

            //Calculando a quantidade de latas necessárias
            quantidade_latas = quantidade_tinta/2 ;

            //Procedimento para arredondar a quantidade de latas
            quantidade_latas = (double)Math.Ceiling(quantidade_latas);
            
            //Mostrar ao usuário a quantidade de latas
            Console.WriteLine("Você usará {0}L de tinta e precisará de {1} lata/as", quantidade_tinta, quantidade_latas);
        }
    }
}
```

### Exercício 5: Quantidade de tijolos ###

image.png

```python
%%writefile tijolos.cs

using System;

namespace myApp
{
    class Program
    {
        static void Main()
        {
        //Preparando variáveis
        float   Largura_Dos_Tijolos, 
                Altura_Dos_Tijolos, 
                Largura_Da_Porta, 
                Altura_Da_Porta, 
                Largura_Da_Janela, 
                Altura_Da_Janela, 
                Largura_Da_Parede,
                Altura_Da_Parede, 
                Area_Dos_Tijolos, 
                Area_Da_Porta, 
                Area_Da_Janela, 
                Area_Da_Parede, 
                Quantidade_De_Tijolos;
        
        // Pegando o valor das variáveis

        Console.WriteLine("Para que o programa funcione bem utilize a mesma unidade de medida, seja ela centimétros ou metros");

        Console.WriteLine("Digite a largura, pressione enter, depois a altura dos tijolos usados e pressione enter");
        Largura_Dos_Tijolos = float.Parse(Console.ReadLine());
        Altura_Dos_Tijolos  = float.Parse(Console.ReadLine());

        Console.WriteLine("Repita o processoa acima para as medidas da porta");
        Largura_Da_Porta    = float.Parse(Console.ReadLine());
        Altura_Da_Porta     = float.Parse(Console.ReadLine());

        Console.WriteLine("Agora faça o mesmo para as medidas da janela");
        Largura_Da_Janela   = float.Parse(Console.ReadLine());
        Altura_Da_Janela    = float.Parse(Console.ReadLine());

        Console.WriteLine("E por ultimo, repita o processo para as medidas da parede");
        Largura_Da_Parede   = float.Parse(Console.ReadLine());
        Altura_Da_Parede    = float.Parse(Console.ReadLine());

        // Processamento dos dados recolhidos

        //Cálculo das áreas

        Area_Dos_Tijolos  = Largura_Dos_Tijolos * Altura_Dos_Tijolos;

        Area_Da_Porta     = Largura_Da_Porta    * Altura_Da_Porta;

        Area_Da_Janela    = Largura_Da_Janela   * Altura_Da_Janela;

        Area_Da_Parede    = ( Largura_Da_Parede  * Altura_Da_Parede * 4 ) - ( Area_Da_Porta + Area_Da_Janela );

        //Calculando a quantidade de tijolos

        Quantidade_De_Tijolos = Area_Da_Parede / Area_Dos_Tijolos;

        Quantidade_De_Tijolos = (float)Math.Ceiling(Quantidade_De_Tijolos);

        //Exibindo os resultados

        Console.WriteLine("A sala tem uma área de {0}U.A, logo serão necessários {1} tijolos", Area_Da_Parede, Quantidade_De_Tijolos);

        }
    }
}
```

```python
%%writefile soma_dentro_de_um_intervalo.cs

using System;

namespace myApp
{
    class Program
    {
        static void Main()
        {
            // Setando variáveis
            int inicioIntervalo, fimIntervalo, soma, valorAtual;

            // Pegando os dados com o usuário
            Console.WriteLine("Digite o inicio do intervalo: ");
            inicioIntervalo = int.Parse(Console.ReadLine());

            Console.WriteLine("Digite o fim do intervalo: ");
            fimIntervalo = int.Parse(Console.ReadLine());

            // Sequência Crescente
            if ( inicioIntervalo < fimIntervalo )
            {
                valorAtual = inicioIntervalo;
                soma = 0;

                while (valorAtual < (fimIntervalo + 1))
                {
                    if ( (valorAtual % 2) != 0 )
                    {
                        soma = soma + valorAtual;
                    }
                    valorAtual++;
                }

            }
            //sequência decrescente
            else 
            {
                valorAtual = inicioIntervalo;
                soma = 0;

                while (valorAtual > (fimIntervalo - 1))
                {
                    if ( (valorAtual % 2) != 0 )
                    {
                        soma = soma + valorAtual;
                    }
                    valorAtual--;
                }
            }
            
            //Exibindo o resultado
            Console.WriteLine(soma);
        }
    }
}
```

### Matriz triângular ###

```python
%%writefile MatrizTriangular.cs

using System;

namespace myApp
{
    class Program
    {
        static void Main()
        {
            // Setando variáveis necessárias
            int ordemMatriz, i, j;

            // Explicando o funcionamento do programa para o usuário
            Console.WriteLine("Olá usuário, este é um programa para dada uma ordem de uma matriz, eu lhe devolvo uma matriz triângular superior com valores aleatórios ");

            // Pegando o valor da ordem da matriz
            Console.WriteLine("Para começarmos, por gentileza, digite a ordem da matriz ( lembrando que para ser triangular superior, ela será uma matriz quadrada )");
            ordemMatriz = int.Parse(Console.ReadLine());

            // Setando a matriz
            int[,] mts = new int [ordemMatriz, ordemMatriz];

            // Preparações para a seleção de numeros randomicos
            Random numAleatorio = new Random();

            // Preenchendo a matriz com números aleatórios

            for (i = 0; i < ordemMatriz; i++)
            {
              for (j = 0; j < ordemMatriz; j++)
              {
                if (j >= i)
                {
                    int valorInteiro = numAleatorio.Next(0,99);
                    mts [i,j] = valorInteiro;
                }else 
                {
                    mts [i,j] = 0;
                }
              }  
            }

            // Exibindo a matriz para o usuário
            for (i = 0; i < ordemMatriz; i++)
            {
              for (j = 0; j < ordemMatriz; j++)
              {
                Console.Write("{0}\t",mts[i,j]);
              }  
              Console.WriteLine("\n");
            }
        }
    }
}
```

```python
%%writefile Apostas.cs

using System;

namespace myApp
{
    class Program 
    {
        static void Main()
        {
            //--- Setando variáveis necessárias
            int i = 0, j = 0, decisao = 0,  resultado = 0, pontos = 0, quantidadeDeTimes = 0;
            string nomeTimes;

            //--- Explicando ao usuário o funcionamento do programa
            Console.WriteLine("Este programa permite a uma pessoa apostar em uma série jogos e verificar o resultado dessas apostas");

            //--- Pegando a quantidade de times que participarão da aposta com o usuário
            Console.WriteLine("Por favor, digite a quantidade de times que participarão da aposta:\n");
            quantidadeDeTimes = int.Parse(Console.ReadLine());

            //--- Criando as matrizes que serão usadas
            string [,] matrizApostas = new string [quantidadeDeTimes,4];
            string [,] matrizResultados = new string [quantidadeDeTimes,4];

            //--- Pegando os nomes dos time 
            //--- Decisão de se o usuário escreverá ou não o nome dos times 
            Console.WriteLine("\nVocê deseja informar os nomes dos times? ( 1 para 'sim', 2 para 'não' )");
            decisao = int.Parse(Console.ReadLine());

            //--- Usuário escolheu escrever o nome dos times 
            if (decisao == 1)
            {
                for (i = 0; i < quantidadeDeTimes; i++)
                {
                    Console.WriteLine("Times da partida {0}", i+1);
                    nomeTimes = Console.ReadLine();
                    matrizApostas[i, 0] = nomeTimes;

                }
            }else //--- Usuário deixa a máquina setar o nome dos times 
            {
                for (i = 0; i < quantidadeDeTimes; i++)
                {
                    nomeTimes = "time " + (i + 1);
                    matrizApostas[i, 0] = nomeTimes;
                }
            }
            
            //--- Pegando as das apostas do usuário
            Console.WriteLine("\nDigite a sua(s) aposta(s) para os resultados dos jogos");

            //___________________________________________________________________________________________________________________
            for ( i = 0; i < quantidadeDeTimes ; i++ )
            {
                Console.WriteLine("\n{0}\n\n1 - vitória\t2 - empate\t3 - derrota\t4 - vitoria e empate\t5 - vitoria e derrota\t6 - empate e derrota",matrizApostas[i, 0]);
                resultado = int.Parse(Console.ReadLine());

                switch (resultado)
                {
                    case 1:
                    matrizApostas[i, 1] = "1";
                    matrizApostas[i, 2] = "0";
                    matrizApostas[i, 3] = "0";                    
                    break;

                    case 2: 
                    matrizApostas[i, 1] = "0";
                    matrizApostas[i, 2] = "1";
                    matrizApostas[i, 3] = "0";
                    break;

                    case 3: 
                    matrizApostas[i, 1] = "0";
                    matrizApostas[i, 2] = "0";
                    matrizApostas[i, 3] = "1";
                    break;

                    case 4: 
                    matrizApostas[i, 1] = "1";
                    matrizApostas[i, 2] = "1";
                    matrizApostas[i, 3] = "0";
                    break;

                    case 5: 
                    matrizApostas[i, 1] = "1";
                    matrizApostas[i, 2] = "0";
                    matrizApostas[i, 3] = "1";
                    break;

                    case 6: 
                    matrizApostas[i, 1] = "0";
                    matrizApostas[i, 2] = "1";
                    matrizApostas[i, 3] = "1";
                    break;

                    default:
                    Console.WriteLine("Resultado inválido, por favor, digite uma opção válida");
                    break;
                }
            }

            //--- Resetando variáveis já usadas que serão úteis nessa etapa
            i = 0;            

            //___________________________________________________________________________________________________________________

            //--- Pegando os resultados das apostas com o usuário
            Console.WriteLine("\nPor favor, digite o gabarito dos resultados dos jogos");
            
            while ( i < quantidadeDeTimes )
            {
                matrizResultados[i,0] = matrizApostas[i,0];// pegando os nomes dos times da matriz de aposta para a de resultados
                
                //--- Pegando os resultados
                Console.WriteLine("\n{0}\n\n1 - vitória\t2 - empate\t3 - derrota",matrizResultados[i, 0]);
                resultado = int.Parse(Console.ReadLine());

                switch (resultado)
                {
                    case 1: 
                    matrizResultados[i, 1] = "1";
                    matrizResultados[i, 2] = "0";
                    matrizResultados[i, 3] = "0";
                    i++;
                    break;

                    case 2: 
                    matrizResultados[i, 1] = "0";
                    matrizResultados[i, 2] = "1";
                    matrizResultados[i, 3] = "0";
                    i++;
                    break;

                    case 3: 
                    matrizResultados[i, 1] = "0";
                    matrizResultados[i, 2] = "0";
                    matrizResultados[i, 3] = "1";
                    i++;
                    break;

                    default:
                    Console.WriteLine("Resultado inválido, por favor, digite uma opção válida");
                    break;
                }
            }

            //___________________________________________________________________________________________________________________
            Console.WriteLine("\nComparando resultado...\n");
            
            //--- Comparando a aposta com o gabarito, percorrer a matriz, e se a aposta for correta, aumenta a pontuação do usuário 
            for ( i = 0; i < quantidadeDeTimes ; i++ )
            {
                for( j = 1 ; j < 4 ; j++ )
                {
                   if ( matrizApostas[i,j].Equals(matrizResultados[i,j]) && matrizResultados[i,j] == "1")
                   {
                       pontos++;
                   }
                }
            }           
            
            Console.WriteLine("\nAs apostas foram :\n");

            //--- Exibição da matriz de apostas para o usuário
            for (i = 0; i < quantidadeDeTimes; i++)
            {
                for (j = 0; j < 4; j++){
                    Console.Write("{0}\t",matrizApostas[i, j]);
                    
                }
                Console.WriteLine("\n");
            }

            //___________________________________________________________________________________________________________________
            Console.WriteLine("\nE os resultados foram :\n");

            //--- Exibição da matriz de gabarito para o usuário
            for (i = 0; i < quantidadeDeTimes; i++)
            {
                for (j = 0; j < 4; j++){
                    Console.Write("{0}\t",matrizResultados[i, j]);
                    
                }
                Console.WriteLine("\n");
            }

            Console.WriteLine("\nPortanto o jogador fez {0} pontos\n", pontos);
        }
    }
}
```
