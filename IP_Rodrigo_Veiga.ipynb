### Exercício 3: Quantidade de latas ###

image.png

```python
%%writefile Latas.cs

using System;

namespace myApp 
{
    class Program 
    {
        static void Main()
        {
            //Declarando variáveis
            double largura_parede, altura_parede, area_parede, quantidade_tinta, quantidade_latas;

            //Pegando os dados da parede com o usuário
            Console.WriteLine("Por favor, digite a altura e largura da parede a ser pintada");
            altura_parede  = double.Parse(Console.ReadLine());
            largura_parede = double.Parse(Console.ReadLine());

            //Calculando a área
            area_parede = altura_parede * largura_parede;

            //Calculando quantidade de tinta em litros
            quantidade_tinta = area_parede * 0.3 ;

            //Calculando a quantidade de latas necessárias
            quantidade_latas = quantidade_tinta/2 ;

            //Procedimento para arredondar a quantidade de latas
            quantidade_latas = (double)Math.Ceiling(quantidade_latas);
            
            //Mostrar ao usuário a quantidade de latas
            Console.WriteLine("Você usará {0}L de tinta e precisará de {1} lata/as", quantidade_tinta, quantidade_latas);
        }
    }
}
```

### Exercício 4: tempo ###

```python
%%writefile tempo.cs

using System;

namespace myApp
{
    class Program
    {
        static void Main()
        {
            // Alocando variáveis
            int Horas, Minutos, Segundos, Total_De_Segundos;

            //Pegando os valores correspondente as Horas, minutos e segundos
            Console.WriteLine("Esse programa converte um horário no formato ( horas-minutos-segundos) em quantidade de segundos");
            Console.WriteLine("Digite o número de horas e aperte enter, repita o processo para minutos e segundos.");
            Horas    = int.Parse(Console.ReadLine());
            Minutos  = int.Parse(Console.ReadLine());
            Segundos = int.Parse(Console.ReadLine());

            //Conversão de formato H/M/S para segundos
            Total_De_Segundos = ( Horas * 3600 ) + ( Minutos * 60 ) + Segundos ;

            //Retornando o resultado ao usuário
            Console.WriteLine("O tempo passado em segundos foi {0}s", Total_De_Segundos);
        }
    }
}
```

### Exercício 5: Quantidade de tijolos ###

image.png

```python
%%writefile tijolos.cs

using System;

namespace myApp
{
    class Program
    {
        static void Main()
        {
        //Preparando variáveis
        float   Largura_Dos_Tijolos, 
                Altura_Dos_Tijolos, 
                Largura_Da_Porta, 
                Altura_Da_Porta, 
                Largura_Da_Janela, 
                Altura_Da_Janela, 
                Largura_Da_Parede,
                Altura_Da_Parede, 
                Area_Dos_Tijolos, 
                Area_Da_Porta, 
                Area_Da_Janela, 
                Area_Da_Parede, 
                Quantidade_De_Tijolos;
        
        // Pegando o valor das variáveis

        Console.WriteLine("Para que o programa funcione bem utilize a mesma unidade de medida, seja ela centimétros ou metros");

        Console.WriteLine("Digite a largura, pressione enter, depois a altura dos tijolos usados e pressione enter");
        Largura_Dos_Tijolos = float.Parse(Console.ReadLine());
        Altura_Dos_Tijolos  = float.Parse(Console.ReadLine());

        Console.WriteLine("Repita o processoa acima para as medidas da porta");
        Largura_Da_Porta    = float.Parse(Console.ReadLine());
        Altura_Da_Porta     = float.Parse(Console.ReadLine());

        Console.WriteLine("Agora faça o mesmo para as medidas da janela");
        Largura_Da_Janela   = float.Parse(Console.ReadLine());
        Altura_Da_Janela    = float.Parse(Console.ReadLine());

        Console.WriteLine("E por ultimo, repita o processo para as medidas da parede");
        Largura_Da_Parede   = float.Parse(Console.ReadLine());
        Altura_Da_Parede    = float.Parse(Console.ReadLine());

        // Processamento dos dados recolhidos

        //Cálculo das áreas

        Area_Dos_Tijolos  = Largura_Dos_Tijolos * Altura_Dos_Tijolos;

        Area_Da_Porta     = Largura_Da_Porta    * Altura_Da_Porta;

        Area_Da_Janela    = Largura_Da_Janela   * Altura_Da_Janela;

        Area_Da_Parede    = ( Largura_Da_Parede  * Altura_Da_Parede * 4 ) - ( Area_Da_Porta + Area_Da_Janela );

        //Calculando a quantidade de tijolos

        Quantidade_De_Tijolos = Area_Da_Parede / Area_Dos_Tijolos;

        Quantidade_De_Tijolos = (float)Math.Ceiling(Quantidade_De_Tijolos);

        //Exibindo os resultados

        Console.WriteLine("A sala tem uma área de {0}U.A, logo serão necessários {1} tijolos", Area_Da_Parede, Quantidade_De_Tijolos);

        }
    }
}
```

###Exercício 6: Diferença de datas###

image.png

```python
%%writefile diferença_de_datas.cs

using System;

namespace myApp
{
    class Program
    {
        static void Main()
        {
            //Setando as variáveis
            float Ano_1, Mes_1, Dia_1, Ano_2, Mes_2, Dia_2, Processo_a, Processo_b, Processo_c ;

            //Pegando os dados necessários com o usuário
            Console.WriteLine ("Digite um ano, depois um mês (em número), depois um dia, sempre apertando a tecla enter após digita-los ");
            Ano_1 = float.Parse(Console.ReadLine());
            Mes_1 = float.Parse(Console.ReadLine());
            Dia_1 = float.Parse(Console.ReadLine());

            Console.WriteLine ("Repita o mesmo procedimento só que agora com uma data mais recente que a anterior");
            Ano_2 = float.Parse(Console.ReadLine());
            Mes_2 = float.Parse(Console.ReadLine());
            Dia_2 = float.Parse(Console.ReadLine());

            //Processamento das diferenças entre as datas
            Processo_a = ( Ano_2 - Ano_1 ) + ( (Mes_2 - Mes_1 ) / 12) + ((Dia_2 - Dia_1 ) / 360);
            Processo_b = (( Ano_2 - Ano_1 ) * 12) + (Mes_2 - Mes_1 ) + ((Dia_2 - Dia_1 ) / 30);
            Processo_c = (( Ano_2 - Ano_1 ) * 365) + ((Mes_2 - Mes_1 ) * 30 ) + ( Dia_2 - Dia_1 );

            //Mostrando o resultado ao usuário
            Console.WriteLine("A diferença de tempo em anos foi : {0}", Processo_a);
            Console.WriteLine("A diferença de tempo em meses foi : {0}", Processo_b);
            Console.WriteLine("A diferença de tempo em dias foi : {0}", Processo_c);
            }
    }
}
```

```python
%%writefile numero_prop.cs

using System;

namespace myApp
{
    class Program 
    {
        static void Main()
        {
            // Criando variáveis
            double aux, aux2, aux3, aux4, aux5, numero;

            // Pegando o valor com o usuário
            Console.WriteLine("Digite um número inteiro com até quatro algarismos \t");
            numero = double.Parse(Console.ReadLine());

            // Separando o numero de 2 em 2 algarismos
            aux  = numero/100;
            aux  = (int)aux;
            aux2 = aux * 100;
            aux3 = numero - aux2;
            
            // Fazendo a operação correspondente a propriedade do enunciado da questão
            aux4 = aux + aux3;
            aux5 = Math.Pow(aux4, 2);
            Console.WriteLine(aux5);
            
            //Verificando se o valor atendeu a propriedade;
            if (numero == aux5)
            {
                Console.WriteLine("Sim, o número {0} possui a propriedade {1} + {2} = {3} e {3}² = {0}",numero, aux, aux3, aux4 ); //Testar funcionalidade
            }
            else 
            {
                Console.WriteLine("Não, o número {0} não possui a propriedade {1} + {2} = {3} e {3}² = {0} já que {3}² = {4}",numero, aux, aux3, aux4, aux5 ); //Testar funcionalidade
            }
        }
    }
}
```

### Exercício 3: Descobrindo se um numero é par ou ímpar ###

image.png

```python
%%writefile paridade.cs

using System;

namespace myApp
{
    class Program
    {
        static void Main()
        {
            // Setando variáveis
            int numero, paridade;

            // Pegando o valor do número com o usuário
            Console.WriteLine("Digite um número inteiro");
            numero = int.Parse(Console.ReadLine());

            // Processamento do resto da divisão por 2
            paridade = numero % 2;

            // Determinando se o numero é inteiro e exibindo o resultado ao usuário
            if ( paridade == 0 )
            {
                Console.WriteLine("O número {0} é par", numero);
            }
            else
            {
                Console.WriteLine("O número {0} é ímpar", numero);
            }
        }
    }
}
```
