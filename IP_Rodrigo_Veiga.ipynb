### Exercício 3: Quantidade de latas ###

```python
%%writefile Latas.cs

using System;

namespace myApp 
{
    class Program 
    {
        static void Main()
        {
            //Declarando variáveis
            double largura_parede, altura_parede, area_parede, quantidade_tinta, quantidade_latas;

            //Pegando os dados da parede com o usuário
            Console.WriteLine("Por favor, digite a altura e largura da parede a ser pintada");
            altura_parede  = double.Parse(Console.ReadLine());
            largura_parede = double.Parse(Console.ReadLine());

            //Calculando a área
            area_parede = altura_parede * largura_parede;

            //Calculando quantidade de tinta em litros
            quantidade_tinta = area_parede * 0.3 ;

            //Calculando a quantidade de latas necessárias
            quantidade_latas = quantidade_tinta/2 ;

            //Procedimento para arredondar a quantidade de latas
            quantidade_latas = (double)Math.Ceiling(quantidade_latas);
            
            //Mostrar ao usuário a quantidade de latas
            Console.WriteLine("Você usará {0}L de tinta e precisará de {1} lata/as", quantidade_tinta, quantidade_latas);
        }
    }
}
```

### Exercício 5: Quantidade de tijolos ###

image.png

```python
%%writefile tijolos.cs

using System;

namespace myApp
{
    class Program
    {
        static void Main()
        {
        //Preparando variáveis
        float   Largura_Dos_Tijolos, 
                Altura_Dos_Tijolos, 
                Largura_Da_Porta, 
                Altura_Da_Porta, 
                Largura_Da_Janela, 
                Altura_Da_Janela, 
                Largura_Da_Parede,
                Altura_Da_Parede, 
                Area_Dos_Tijolos, 
                Area_Da_Porta, 
                Area_Da_Janela, 
                Area_Da_Parede, 
                Quantidade_De_Tijolos;
        
        // Pegando o valor das variáveis

        Console.WriteLine("Para que o programa funcione bem utilize a mesma unidade de medida, seja ela centimétros ou metros");

        Console.WriteLine("Digite a largura, pressione enter, depois a altura dos tijolos usados e pressione enter");
        Largura_Dos_Tijolos = float.Parse(Console.ReadLine());
        Altura_Dos_Tijolos  = float.Parse(Console.ReadLine());

        Console.WriteLine("Repita o processoa acima para as medidas da porta");
        Largura_Da_Porta    = float.Parse(Console.ReadLine());
        Altura_Da_Porta     = float.Parse(Console.ReadLine());

        Console.WriteLine("Agora faça o mesmo para as medidas da janela");
        Largura_Da_Janela   = float.Parse(Console.ReadLine());
        Altura_Da_Janela    = float.Parse(Console.ReadLine());

        Console.WriteLine("E por ultimo, repita o processo para as medidas da parede");
        Largura_Da_Parede   = float.Parse(Console.ReadLine());
        Altura_Da_Parede    = float.Parse(Console.ReadLine());

        // Processamento dos dados recolhidos

        //Cálculo das áreas

        Area_Dos_Tijolos  = Largura_Dos_Tijolos * Altura_Dos_Tijolos;

        Area_Da_Porta     = Largura_Da_Porta    * Altura_Da_Porta;

        Area_Da_Janela    = Largura_Da_Janela   * Altura_Da_Janela;

        Area_Da_Parede    = ( Largura_Da_Parede  * Altura_Da_Parede * 4 ) - ( Area_Da_Porta + Area_Da_Janela );

        //Calculando a quantidade de tijolos

        Quantidade_De_Tijolos = Area_Da_Parede / Area_Dos_Tijolos;

        Quantidade_De_Tijolos = (float)Math.Ceiling(Quantidade_De_Tijolos);

        //Exibindo os resultados

        Console.WriteLine("A sala tem uma área de {0}U.A, logo serão necessários {1} tijolos", Area_Da_Parede, Quantidade_De_Tijolos);

        }
    }
}
```

### Exercício 9: inversoDinheiro ###

image.png

```python
%%writefile InversoDinheiro.cs

using System;

namespace myApp 
{
    class Program
    {
        static void Main()
        {
            // Setando variáveis necessárias
            int notas200 = 0, notas100 = 0, notas50 = 0, notas20 = 0, notas10 = 0, notas5 = 0, notas2 = 0, notas1 = 0;
            float totalDinheiro, op, aux1;

            // Pegando o valor total do dinheiro com o usuário
            Console.WriteLine("Digite o montante de dinheiro ( esse programa náo é compatível com moedas)");
            totalDinheiro = int.Parse(Console.ReadLine());

            // Convertendo o valor para notas menores

            // notas de 200
            op = totalDinheiro/200;
            notas200 = (int)op;
            aux1 = op - (float)notas200;
            aux1 = aux1 * 200;
            
            
            if (aux1 > 0)
            {
                //notas de 100
                totalDinheiro = aux1;
                op = totalDinheiro/100;
                notas100 = (int)op;
                aux1 = op - notas100;
                aux1 = aux1 * 100;
            } 
            
            if (aux1 != 0)
            {
                //notas de 50
                totalDinheiro = aux1;
                op = totalDinheiro/50;
                notas50 = (int)op;
                aux1 = op - notas50;
                aux1 = aux1 * 50;
            }
            
            if (aux1 != 0)
            {
                //notas de 20
                totalDinheiro = aux1;
                op = totalDinheiro/20;
                notas20 = (int)op;
                aux1 = op - notas20;
                aux1 = aux1 * 20;
            }
            
            if (aux1 != 0)
            {
                //notas de 10
                totalDinheiro = aux1;
                op = totalDinheiro/10;
                notas10 = (int)op;
                aux1 = op - notas10;
                aux1 = aux1 * 10;
            }
            
            if (aux1 != 0)
            {
                //notas de 5
                totalDinheiro = aux1;
                op = totalDinheiro/5;
                notas5 = (int)op;
                aux1 = op - notas5;
                aux1 = aux1 * 5;
            }
            
            if (aux1 != 0)
            {
                //notas de 2
                totalDinheiro = aux1;
                op = totalDinheiro/2;
                notas2 = (int)op;
                aux1 = op - notas2;
                aux1 = aux1 * 2;
                notas1 = (int)aux1;
            }
            
            Console.WriteLine("O resultado é \n{0} notas de 200\n{1} notas de 100\n{2} notas de 50\n{3} notas de 20\n{4} notas de 10\n{5} notas de 5\n{6} notas de 2\n{7} notas de 1", notas200, notas100, notas50, notas20, notas10, notas5, notas2, notas1);
        }
    }
}
```

### Exercício 10: Corrida ###

Questão 10: Um autódromo necessita de um programa para
verificar qual o carro mais rápido em uma competição.
Considerando que dois carros participam da corrida, o programa pergunta qual o comprimento da pista (em metros) e quanto tempo (em segundos) cada um dos carros levou pra percorrer a pista, e informa a velocidade do carro mais rápido. (Considere que em uma corrida não existem dois carros com tempos de percurso exatamente iguais).

image.png

```python
%%writefile corrida.cs

using System;

namespace myApp
{
    class Program 
    {
        static void Main()
        {
            // Setando as variáveis necessárias
            int tempoCarro1, tempoCarro2;
            float comprimentoPista, velocidade;

            // Explicando ao usuário o programa
            Console.WriteLine("Olá, esse programa lhe dirá qual dos dois carros chegou mais rapido a linha de chegada e qual foi a velocidade, para comecarmos, por favor, digite o comprimento da pista em m");
            
            // Pegando o comprimento da pista 
            comprimentoPista = float.Parse(Console.ReadLine());

            // Tempo do carro 1
            Console.WriteLine("Agora digite quanto tempo em segundos o carro 1 demorou para completar o circuito");
            tempoCarro1 = int.Parse(Console.ReadLine());

            // Tempo do carro 2
            Console.WriteLine("Agora digite quanto tempo em segundos o carro 2 demorou para completar o circuito");
            tempoCarro2 = int.Parse(Console.ReadLine());
            
            //vendo qual carro chegou primeiro

            if( tempoCarro1< tempoCarro2)
            {
                velocidade = comprimentoPista/tempoCarro1;
                Console.WriteLine("O carro 1 venceu!!! E ele estava a {0} m/s", velocidade);
            }
            else 
            {
                velocidade = comprimentoPista/tempoCarro1;
                Console.WriteLine("O carro 2 venceu!!! E ele estava a {0} m/s", velocidade);
            }
        }
    }
}
```

### Exercício 1: PA ###

image.png

```python
%%writefile soma_elementos_iniciais_pa.cs

using System;

namespace myApp
{
    class Program
    {
        static void Main()
        {
            //Setando variáveis
            int i = 0;
            float valorInicial, valorAtual, numeroElementos, razao, soma = 0;

            //Explicando o funcionamento do programa
            Console.WriteLine("Esse programa soma os primeiros 'n' valores de uma PA");
            
            //Pedindo os dados da PA ao usuário
            Console.WriteLine("Digite o valor inicial, o numero de elementos, e a razão (nessa ordem ) da progressão aritimética desejada");
            
            //pegando o valor inicial da PA
            valorInicial = float.Parse(Console.ReadLine()); 
            
            //pegando o numero de elementos da PA
            numeroElementos = float.Parse(Console.ReadLine());
            
            //pegando a razão da PA
            razao = float.Parse(Console.ReadLine());

            // Processamento
            while (i < numeroElementos)
            {
                i++;
                valorAtual = valorInicial + (i - 1) * razao;
                soma = soma + valorAtual;
            }
            //Mostrando o resultado ao usuário
            Console.WriteLine("A soma dos {0} primeiros elementos é {1}", numeroElementos, soma);
        }
    }
}
```

### Exercício 2: numeros primos ###

image.png

BO's do código

*   Imprime "1" que nesse caso é um falso primo( tem q começar em "2")
*   Divisores = 0 ( é "==" )
*   Programa diferente do enunciado
*   Por que tem q ser d<=n-1


```python
%%writefile numeros_primos.cs

using System;

namespace myApp
{
    class Program
    {
        static void Main()
        {
            //Setando variáveis usadas no programa
            int n,d,numero,divisores;

            //Pegando o numero qual o programa exibirá todos os primos até ele
            Console.WriteLine("Digite um número que eu lhe devolvo todos os numeros primos até ele");
            numero=int.Parse(Console.ReadLine());

            // Contador relacionado a impressão de numeros primos
            for(n=2; n<=numero; n++)
            {
                divisores = 0;

                for(d=2; d<=n-1; d++)
                {
                    if ( n % d == 0 )
                    {
                        divisores++;
                    }
                }

                if (divisores == 0)
                {
                    Console.WriteLine(n);
                }
            }
        }
    }
}
```
